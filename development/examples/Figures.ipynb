```python
# Draw N samples from a bivariate normal distribution with true correlation 0.8. 
# Repeat this 1000 times. Let N be 10, 50 and 100. 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
corr_10 = []
corr_50 = []
corr_100 = []
for i in range(1000):
    mean = [0, 0]  # mean values
    cov = [[1, 0.4], [0.4, 1]]  # covariance matrix
    samples = np.random.multivariate_normal(mean, cov, size=10)
    corr_10.append(pearsonr(samples[:,0], samples[:,1])[0])
    samples = np.random.multivariate_normal(mean, cov, size=50)
    corr_50.append(pearsonr(samples[:,0], samples[:,1])[0])
    samples = np.random.multivariate_normal(mean, cov, size=100)
    corr_100.append(pearsonr(samples[:,0], samples[:,1])[0])
```

```python
# Plot the histogram of the Pearson correlation coefficient in a 1 x 3 subplot.
# Let the histogram be light blue with black borders.
# Let the size of the figure be 10 x 8 inches
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].hist(corr_10, bins=20, color='lightblue', edgecolor='black', label='Distribution of sample correlations')
ax[0].set_title('r = 0.4, sample size n = 10')
# Mark the average sample correlation coefficient with a round blue marker
# and the true correlation with a round red marker. Let the markers overlap with
# the axis.
ax[0].plot(np.mean(corr_10), 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='blue', markerfacecolor='none', clip_on=False, label='Average sample correlation')
ax[0].plot(0.4, 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='red', markerfacecolor='none', clip_on=False, label='True correlation')
ax[0].set_xlim([-0.7, 1])
ax[0].legend()
ax[0].set_xlabel('sample correlation')
ax[0].set_ylabel('frequency')
ax[1].hist(corr_50, bins=20, color='lightblue', edgecolor='black')
ax[1].set_title('r = 0.4, sample size n = 50')
ax[1].plot(np.mean(corr_50), 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='blue', markerfacecolor='none', clip_on=False)
ax[1].plot(0.4, 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='red', markerfacecolor='none', clip_on=False)
ax[1].set_xlim([-0.7, 1])
ax[1].set_xlabel('sample correlation')
ax[2].hist(corr_100, bins=20, color='lightblue', edgecolor='black')
ax[2].set_title('r = 0.4, sample size n = 100')
ax[2].plot(np.mean(corr_100), 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='blue', markerfacecolor='none', clip_on=False)
ax[2].plot(0.4, 0, marker='o', markersize=8, markeredgewidth=2, markeredgecolor='red', markerfacecolor='none', clip_on=False)
ax[2].set_xlim([-0.7, 1])
ax[2].set_xlabel('sample correlation')
fig.tight_layout()
plt.savefig('r04.pdf', dpi=300)
plt.show()
```
